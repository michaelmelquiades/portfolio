# Try to make a feature/target dataset

# Try without any treatment of the data first

```python
data = dict(sample = [], 
            mean = [], std = [],
            maxi = [], mini = [], 
            wave_maxi = [], wave_mini = [])
for i in ref.id:
    s = str(i)
    data['sample'] += [s]
    data['mean'] += [df[s].mean()]
    data['std'] += [df[s].std()]
    data['maxi'] += [df[s].max()]
    data['mini'] += [df[s].min()]
    data['wave_maxi'] += [df[df[s] == df[s].max()].wavelength.values[0]]
    data['wave_mini'] += [df[df[s] == df[s].min()].wavelength.values[0]]
```

```python
df_poss = pd.DataFrame(data)
df_poss['cp'] = ref.CP
df_poss['ndf'] = ref.NDF
df_poss
```

# Try some simple regression techniques

```python
from sklearn.linear_model import LinearRegression
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import train_test_split, RepeatedKFold, cross_val_score
```

```python
x = df_poss.iloc[:, 1:5]
y = df_poss.iloc[:, 7:]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
lin = MultiOutputRegressor(LinearRegression())
```

```python
cv = RepeatedKFold(n_splits = 10, n_repeats = 5, random_state=42)
scores = cross_val_score(lin, x, y, scoring='neg_mean_absolute_error', cv = cv, n_jobs = -1)
scores = np.absolute(scores)
print(f'MAE: {np.mean(scores)} ({np.std(scores)})')
```

# Technically better, though requires more work to get it to do multi. Not particularly a problem, but is it worth it for a <2% decrease in error?

```python
from sklearn.model_selection import GridSearchCV
from bayes_opt import BayesianOptimization
from sklearn.metrics import classification_report, mean_absolute_error, precision_score, recall_score, f1_score
```

```python
grid_params = {'n_estimators':[10, 30, 50, 100, 200, 500, 1000], 'criterion':['friedman_mse', 'poisson', 'absolute_error', 'squared_error']}
bayes_params = {'min_samples_split':(0, 1), 'min_samples_leaf':(0, 0.5)}
```

```python
clf = GridSearchCV(RandomForestRegressor(random_state = 42), param_grid = grid_params)
clf.fit(x_train, y_train)

def tree_func(min_samples_split, min_samples_leaf):
    tree = RandomForestRegressor(criterion=clf.best_params_['criterion'], 
                                  n_estimators=clf.best_params_['n_estimators'], 
                                  random_state=42)
    tree.fit(x_train, y_train)
    y_preds = tree.predict(x_test)
    score = mean_absolute_error(y_test, y_preds)
    return score

optimizer = BayesianOptimization(f = tree_func,
                                 pbounds = bayes_params,
                                 random_state = 42)
optimizer.maximize()

rfr = RandomForestRegressor(criterion = clf.best_params_['criterion'], 
                             n_estimators = clf.best_params_['n_estimators'], 
                             min_samples_leaf = optimizer.max['params']['min_samples_leaf'], 
                             min_samples_split = optimizer.max['params']['min_samples_split'], 
                             random_state = 42)
rfr.fit(x_train, y_train)
```

```python
cv = RepeatedKFold(n_splits = 10, n_repeats = 5, random_state=42)
scores = cross_val_score(rfr, x_test, y_test, scoring='neg_mean_absolute_error', cv = cv, n_jobs = -1)
scores = np.absolute(scores)
print(f'MAE: {np.mean(scores)} ({np.std(scores)})')
```
