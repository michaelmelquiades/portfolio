```python
post_dfs = {}
for i in files:
    try:
        data = prep_edf(ext+i+'.edf', set_bipolar=True, an_ref=['EEG A1-REF'], cat_ref=['EEG A2-REF'])
    except:
        data = prep_edf(ext+i+'.edf', set_bipolar=True, an_ref=['EEG A1-LE'], cat_ref=['EEG A2-LE'])
        
    post_dfs['df_'+i[5:8]] = csv_maker(data, fname=i, change_rate=True) 
```

```python
post_dfs.keys()
```

```python
df_comp = pd.read_csv('eeg_pges_comp.csv').drop(columns = ['Unnamed: 0', 'T1', 'T2'])
for i in post_dfs.keys():
    tmp = post_dfs[i]
    tmp['labels'] = [0]*len(tmp)
    tmp['file'] = ['supp']*len(tmp)
    tmp_new_cols = [j.lstrip('EEG ') for j in tmp.columns]
    tmp_new_cols = [j.rstrip('-LE') for j in tmp_new_cols]
    tmp_new_cols = [j.rstrip('-REF') for j in tmp_new_cols]
    tmp_cols = dict(zip(tmp.columns, tmp_new_cols))
    tmp = tmp.rename(columns=tmp_cols)
    tmp_red = tmp[df_comp.columns].reset_index(drop=True)
    post_dfs[i] = tmp_red
```

```python
post_dfs['df_biw'] = post_dfs['df_biw'].iloc[153:184, :]
post_dfs['df_ipg'] = post_dfs['df_ipg'].iloc[133:218, :]
post_dfs['df_oya'] = post_dfs['df_oya'].iloc[422:488, :]
post_dfs['df_pas'] = post_dfs['df_pas'].iloc[78:134, :]
```

```python
for i in df_comp.index:
    print(i, df_comp.file[i])
```
